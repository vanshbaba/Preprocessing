```python
#data[data.columns[-3:]] = data[data.columns[-3:]].replace('[$ /year+ annual bonus (u...pt15%f),h]','',regex=True).astype(str)
vansh9 = vansh2[vansh2['location'].str.len() < 40]

def map_yearly_salary_range(val):
    if pd.isnull(val):
        return np.nan
    elif "/year" in val:
        part = val.split("/year")[0].replace("$", " ").strip()
        if "-" in part:
            mn, mx = part.split("-")[0:2]
            try:
                mn = float(mn.replace(",", "").strip())
                mx = float(mx.replace(",", "").strip())
            except:
                return np.nan
            return mn, mx
```

```python
vansh9
```

```python
def map_hourly_salary_range(val):
    if pd.isnull(val):
        return np.nan
    elif "/hour" in val:
        part = val.split("/hour")[0].replace("$", " ").strip()
        if "-" in part:
            mn, mx = part.split("-")[0:2]
            try:
                mn = float(mn.replace(",", "").strip())
                mx = float(mx.replace(",", "").strip())
            except:
                return np.nan
            return mn, mx
        
vansh2 = vansh2.assign(yearly_salary_range=vansh2['salary'].map(map_yearly_salary_range),
                   hourly_salary_range=vansh2['salary'].map(map_hourly_salary_range))
```
